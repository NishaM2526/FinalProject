```python
df = pd.read_csv(r"C:\Users\nisha\OneDrive\Desktop\Nisha\GUVI\FinalProject\BankTermDepositPrediction\Data\bank-full.csv", sep=';')
df_copy = df.copy()

df_copy.drop('duration', axis=1, inplace=True)

# *******ENCODING for categorical variables*********
# One-hot encoding for 'job', 'marital', 'contact', 'month', 'poutcome' columns
df_copy = pd.get_dummies(df_copy,columns=['job','marital','contact','month','poutcome'],dtype='int',drop_first=True)

# Ordinal encoding for 'education' column
maping = {'unknown': 0, 'primary': 1, 'tertiary': 2, 'secondary': 3}
df_copy['education'] = df_copy['education'].map(maping)

# Label encoding for 'default', 'housing', 'loan', 'y' columns
label_encoding_cols = ['default','housing','loan', 'y']
encoder = LabelEncoder()
for col in label_encoding_cols:
    df_copy[col] = encoder.fit_transform(df_copy[col])

# ******* FEATURE ENGINNERING *********
df_copy['age_balance'] = df_copy['age'] * df_copy['balance']
df_copy['contacted_before'] = np.where(df_copy['previous'] > 0, 1, 0)
df_copy['total_loans'] = (df_copy['housing']+ df_copy['loan'])

# ******* FEATURES & TARGET VARIABLE SEPARATION *********
X = df_copy.drop('y', axis=1)
y = df_copy['y']

# ******* Scaling for numerical variables*********
scaler = RobustScaler()
numerical_cols = ['age', 'balance', 'campaign', 'pdays', 'previous']
X_scaled = scaler.fit_transform(df_copy[numerical_cols])

df_copy.head()
```
